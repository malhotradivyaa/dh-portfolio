# Run the portfolio site from a Jupyter cell

This is only needed if you specifically want to launch the site from inside
a notebook. The simpler, more standard way is: open a VS Code terminal in
this folder and run `python app.py` directly.

Running a web server from a notebook cell requires a background thread,
since otherwise the cell would block forever and you couldn't run anything else.

```python
!pip install -q -r requirements.txt
```

```python
import threading
from app import app

def run_app():
    app.run(port=5000, debug=False, use_reloader=False)

thread = threading.Thread(target=run_app, daemon=True)
thread.start()
print("Site running at http://127.0.0.1:5000 — open it in your browser.")
```

Edit `app.py`'s `SITE` dictionary to change your content, save the file, then
re-run the cell above to restart the server with your changes.
